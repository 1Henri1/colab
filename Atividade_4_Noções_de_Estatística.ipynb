### Problema 4: Regressão Linear Simples

Para as três variáveis ("Pontos por Jogo", "Assistências por Jogo" e "Rebotes por Jogo"), gere um gráfico de dispersão para cada e insira uma reta representando a regressão linear simples, com relação ao "valor de marcado".

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("https://drive.google.com/u/3/uc?id=1ilNCHv5ARo_noznV_EWsOr8FaL1rO5d7&export=download")
x1 = df['Pontos por Jogo']
x2 = df['Assistências por Jogo']
x3 = df['Rebotes por Jogo']
y = df['Valor de Mercado (USD)']
(a1, b1) = np.polyfit(x1, y, deg = 1)
(a2, b2) = np.polyfit(x2, y, deg = 1)
(a3, b3) = np.polyfit(x3, y, deg = 1)
y1 = a1*x1+b1
y2 = a2*x2+b2
y3 = a3*x3+b3
plt.plot(x1, y1, 'r-', label = 'Pontos por Jogo')
plt.plot(x2,y2, 'g.-', label = 'Assistências por Jogo')
plt.plot(x3, y3, 'b.-', label = 'Rebotes por Jogo')
plt.legend()
```
